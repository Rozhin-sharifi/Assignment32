```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')

std=1
N=1000

bananas_width= np.random.normal(4,std,N)
bananas_length = np.random.normal(8,std,N)
bananas_weight = np.random.normal(0.5,0.1,N)

apples_width = np.random.normal(6,std,N)
apples_length = np.random.normal(6,std,N)
apples_weight = np.random.normal(0.25,0.1,N)

width=np.concatenate((bananas_width,apples_width))
length=np.concatenate((bananas_length,apples_length))
weight=np.concatenate((bananas_weight,apples_weight))

x=np.array([width,length,weight]).T

bananas_lable=np.zeros(N,dtype='int')
apples_lable=np.ones(N,dtype='int')
y=np.concatenate((bananas_lable,apples_lable))

ax.scatter3D(x[0:,0], x[0:,1], x[0:,2],c=y*0.1, cmap='YlOrRd')
ax.set_xlabel('WIDTH', fontweight ='bold')
ax.set_ylabel('LENGHT', fontweight ='bold')
ax.set_zlabel('WEIGHT', fontweight ='bold')
plt.show()
```
